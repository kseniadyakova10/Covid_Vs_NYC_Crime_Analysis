```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load and read data
data = "../resources/2019data.csv"
data2020 = "../resources/NYPD_Arrest_Data__Year_to_Date_.csv"
nypd_data = pd.read_csv(data, parse_dates=["ARREST_DATE"])
nypd_data2020 = pd.read_csv(data2020, parse_dates=["ARREST_DATE"])
```

```python
#isolate march 2019
march_data_19 = nypd_data.loc[(nypd_data["ARREST_DATE"] > "03/01/2019") & (nypd_data["ARREST_DATE"] < "04/01/2019"), :]

#group by offense type and get count of top 15 offenses in March 2019
group_ofns_19m = march_data_19.groupby("OFNS_DESC")
count_ofns_19m = group_ofns_19m["OFNS_DESC"].count().sort_values(ascending=False).head(15)
count_ofns_19m
```

```python
#isolate March 2020
march_data_20 = nypd_data2020.loc[(nypd_data2020["ARREST_DATE"] > "03/01/2020") & (nypd_data2020["ARREST_DATE"] < "04/01/2020"), :]

#group by offense type and get count of top 15 offenses in March 2020
group_ofns_20m = march_data_20.groupby("OFNS_DESC")
count_ofns_20m = group_ofns_20m["OFNS_DESC"].count().sort_values(ascending=False).head(15)
count_ofns_20m
```

```python
#isolate April 2019
april_data_19 = nypd_data.loc[(nypd_data["ARREST_DATE"] > "04/01/2019") & (nypd_data["ARREST_DATE"] < "05/01/2019"), :]

#group by offense type and get count of top 15 offenses in April 2019
group_ofns_19a = april_data_19.groupby("OFNS_DESC")
count_ofns_19a = group_ofns_19a["OFNS_DESC"].count().sort_values(ascending=False).head(15)
count_ofns_19a
```
