```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_file = "Resources/time_series.csv"
data_file_df = pd.read_csv(data_file)
data_file_df.head()
```

```python
data_file_df[["Year",'Mounth','Day']] = data_file_df["date"].str.split("-",expand=True) 
data_file_df.head()
```

```python
clean_data_file_df = data_file_df[['Mounth','Day','Year','tested','positive','deaths']]
clean_data_file_df.head()
```

```python

Mounth = clean_data_file_df["Mounth"]
positive = clean_data_file_df["positive"]
deaths=clean_data_file_df["deaths"]
tested= clean_data_file_df["tested"]      

#plt.plot(Mounth, positive,"g--", label="Positive")
plt.plot(Mounth, deaths,"r--", label="Deaths")
#plt.plot(Mounth, tested,"y--", label="Tested")


plt.title("Coronavirus cases in New York over time")
plt.xlabel("Mounth")
plt.ylabel("Number of cases")
plt.yticks()
plt.legend()


plt.show()
```
